# Распределение длин ридов

```python
def get_reads_histo_from_fasta(path_to_fasta):
    with open(path_to_fasta, 'r') as fasta:
        lines = fasta.readlines()
        lens = []
        
        current_len = 0
        for i in range(1, len(lines)):
            if lines[i].startswith('>'):
                lens.append(current_len)
                current_len = 0
            else:
                current_len += len(lines[i].strip())
        sns.displot(lens, kde=True)
        
        
path = '/Users/alexandraPrimak/Desktop/fasta.fasta' #вставить сюда путь до нужного файла

get_reads_histo_from_fasta(path)
```
